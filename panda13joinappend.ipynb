#convinient method of combining columns of two potentially different index

#other,on,how,lsuffix,rsuffix

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3],
                  'B':[50,60,70]},index=['a','b','c'])
df2=pd.DataFrame({'C':[4,5],
                  'D':[10,20]},index=['a','b'])
display(df1,df2)
```

```python
df1.join(df2)#value not filled for diff index
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='right')#default left
```

```python
df1.join(df2,how='outer')
```

```python
#to add dataframes of same column use lsuffix/rsuffix
df1=pd.DataFrame({'A':[1,2,3],
                  'B':[50,60,70]},index=['a','b','c'])
df2=pd.DataFrame({'A':[4,5],
                  'D':[10,20]},index=['a','b'])
display(df1,df2)
```

```python
df1.join(df2,lsuffix='_1')
```

```python
df1.join(df2,rsuffix='_1')
```

##append function used to append rows of other dataframe to end of given dataframe,returning a new dataframe object

#other,lgnore_index,sort

```python
df3=pd.DataFrame({'A':[1,2,3],
                  'B':[50,70,89]})
df4=pd.DataFrame({'C':[4,5,6],
                  'B':[40,30,28]})
display(df3,df4)
```

```python
df3.append(df4)
```

```python
df3.append(df4,ignore_index=True)
```

```python
df4.append(df3,ignore_index=True)
```
