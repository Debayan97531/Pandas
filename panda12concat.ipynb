#provides various FACILITIES forbining easily combining together Series,DataFrame,Panel objects

#parametres:objs,axis,join,join_axes,ignore_index,keys,sort

````python
import pandas as pd
````

````python
sr1=pd.Series([0,12,23])
sr1
````

````python
sr2=pd.Series([1,13,24,8,88])
sr2
````

````python
pd.concat([sr1,sr2])
````

````python
df1=pd.DataFrame({'ID':[1,2,3,4],
              'Name':['A','B','C','D'],
              'Class':[6,8,99,7]})
df1
````

````python
df2=pd.DataFrame({'ID':[7,8],
              'Name':['G','H'],
              'Class':[29,37]})
df2
````

````python
pd.concat([df1,df2])
````

````python
pd.concat([df1,df2],axis=1)
````

````python
pd.concat([df2,df1])
````

````python
pd.concat([df1,df2],axis=0,ignore_index=True)#default value false
````

````python
pd.concat([df1,df2],axis=1,join='inner')#default outer
````

````python
pd.concat([df1,df2],axis=1,join_axes=df1.index)
````

````python
df1=pd.DataFrame({'ID':[1,2,3,4],
              'Name':['A','B','C','D'],
              'Class':[6,8,99,7]})
df1
````

````python
df2=pd.DataFrame({'ID':[5,6,7,8],
              'Name':['E','F','G','H'],
              'Class':[4,5,29,37]})
df2
````

````python
pd.concat([df1,df2],keys=['first','ec'])
````

````python
pd.concat([df1,df2],axis=1,keys=['first','ec'])
````
