#replace(to_replace,value)
**dict,regex,method,inplace**

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Fortune_10 (2).csv')
df
```

```python
df.replace(to_replace='Health',value='Finace')
```

```python
df.replace('Health','Finance')#string
```

```python
df.replace(10,20)#int
```

```python
df.replace([1,2,3,4,5,6,7,23],0)
```

```python
df.replace([1,2,3,4,5,6],[11,12,31,32,44,3])
```

```python
df.replace({'Industry':'IT Services'},'none')
```

```python
df.replace({'Industry':'Health','Inception':2009},'none')
```

```python
df.replace('[A-Za-z]',0,regex=True)#regular expression converted all values
```

```python
df.replace({'Industry':'[A-Za-z]'},0,regex=True)#for any particular column
```

```python
df.replace('Health',method='ffill')
```

```python
df.replace('Health',method='bfill')
```

```python
df.replace(2009,method='bfill',limit=2)
```

```python
df.replace(2010,inplace=True)#modify this dataframe
df
```
