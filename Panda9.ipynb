#loc:access a group of rows & columns by label or a bool array

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Fortune_10 (2).csv')
df
```

```python
df.loc[0]#all values of 0th row,given as label
```

```python
df.loc[4]
```

```python
df.loc[[0,1]]#list of labels
```

```python
df.loc[4,'Name']
```

```python
df.loc[0:4,'Name']
```

```python
df.loc[[False,False,True,False,False,True,False,False,True,True]]
```

```python
df.loc[df['ID']<8]
```

```python
df.loc[df['ID']<8,['Industry']]
```

#iloc

```python
df.iloc[0]#input given as int
```

```python
df.iloc[[0]]
```

```python
df.iloc[:,0]#all rows
```

```python
df.iloc[[0,2]]
```
