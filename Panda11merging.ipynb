#merging cloumns or indexes in dataFrame based on one or  more keys

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                 'Class':[9,10,11,12] })
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,5],
                 'Name':['A','B','C','D'] })
df2
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df2,df1,on='ID')#intersection part
```

```python
pd.merge(df1,df2,on='ID',how='inner')
```

```python
pd.merge(df1,df2,on='ID',how='left')#based on df1
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

```python
pd.merge(df1,df2,on='ID',how='outer')#union
```

```python
pd.merge(df1,df2,on='ID',how='outer',indicator=True)
```

```python
df2=pd.DataFrame({'ID':[6,7,8,9],
                 'Name':['A','B','C','D'] })
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)#to show all diff index
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                 'Class':[9,10,11,12] })
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4],
                 'Class':[9,10,11,12] })
df2
```

```python
pd.merge(df1,df2,on='ID')#same value->suffixes work here
```

```python
pd.merge(df1,df2,on='ID',suffixes=('Higher','_Middle'))
```
