#**fillna(mathod:ffill,bfill;axis=0,1,liit,inplace**

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Fortune_10 (2).csv')
df
```

```python
df.fillna(0)#0 in place of NaN
```

```python
df.fillna({'Namw':'none','Growth':1})#using as dict
```

```python
df.fillna(method='ffill')#previous value
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill')#backward value
```

```python
df.fillna(method='ffill',axis=0)#for row
```

```python
df.fillna(method='ffill',axis=1)#value of previous column for Industry
```

```python
df.fillna(method='bfill',axis=1)#see column industry
```

```python
df.fillna(0,limit=2)#just fill 2 value for a column
```

```python
df.fillna(method='ffill',limit=2)
```

```python
df.fillna(5,inplace=True)#5 in place of missing value
df
```
